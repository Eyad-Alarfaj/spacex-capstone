# Data Wrangling and Feature Engineering

Creating target variable and engineering features for predictive modeling.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('spacex_launches_raw.csv')
print(f"Original shape: {df.shape}")
```

### Create Target Variable

```python
# Class is already in the data (1=success, 0=failure)
print(f"Target variable 'Class' created")
print(f"1: {(df['Class']==1).sum()} (Success)")
print(f"0: {(df['Class']==0).sum()} (Failure)")
```

### Feature Engineering

```python
# Flight Experience feature
df['Flight_Experience'] = df['Booster_Landings']

# Payload Class (binning by quantiles)
df['Payload_Class'] = pd.cut(df['Payload_Mass_kg'], bins=3, labels=['Light', 'Medium', 'Heavy'])

print(f"Feature Engineering Complete:")
print(f"- Flight_Experience: 0-5 (from Booster_Landings)")
print(f"- Payload_Class: Light/Medium/Heavy based on quantiles")
```

### One-Hot Encoding

```python
# One-hot encode categorical variables
df = pd.get_dummies(df, columns=['Launch_Site', 'Orbit_Type', 'Payload_Class'], drop_first=False)

print(f"Encoded Launch Sites: CCAFS LC-40, KSC LC-39A, VAFB SLC-4E, Boca Chica")
print(f"Encoded Orbit Types: GEO, HEO, ISS, LEO, SSO")
print(f"Payload Classes: Heavy, Light, Medium")
```

### Select Features for Modeling

```python
# Select relevant features
feature_cols = ['Payload_Mass_kg', 'Flight_Experience'] + \
               [col for col in df.columns if col.startswith('Launch_Site_') or col.startswith('Orbit_Type_') or col.startswith('Payload_Class_')]

X = df[feature_cols]
y = df['Class']

print(f"Final Feature Matrix Shape: {X.shape}")
print(f"Target Variable: Class")
print(f"Features selected for modeling")
```

### Save Processed Data

```python
df.to_csv('spacex_processed.csv', index=False)
print(f"Processed data saved to spacex_processed.csv")
```
