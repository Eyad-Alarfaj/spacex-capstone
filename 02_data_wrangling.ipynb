# Notebook 2 — Data Wrangling

Create the classification target and engineer model-ready features.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('spacex_launches_raw.csv')
print(f'Input shape: {df.shape}')
```

### Target variable
`Class` = 1 when the first stage landed successfully, 0 otherwise.

```python
print(df['Class'].value_counts().rename({1: 'Success (1)', 0: 'Failure (0)'}))
```

### Feature engineering

```python
df['Flight_Experience'] = df['Booster_Landings']
df['Payload_Class'] = pd.cut(df['Payload_Mass_kg'], bins=3, labels=['Light', 'Medium', 'Heavy'])
print(df[['Flight_Experience', 'Payload_Class']].head())
print('\nPayload class counts:')
print(df['Payload_Class'].value_counts())
```

### One-hot encoding of categorical variables

```python
encoded = pd.get_dummies(df, columns=['Launch_Site', 'Orbit_Type', 'Payload_Class'])
feature_cols = ['Payload_Mass_kg', 'Flight_Experience'] + [c for c in encoded.columns if c.startswith(('Launch_Site_', 'Orbit_Type_', 'Payload_Class_'))]
X = encoded[feature_cols]
y = encoded['Class']
print(f'Feature matrix: {X.shape}')
print(f'Features: {list(X.columns)}')
```

### Save processed data

```python
encoded.to_csv('spacex_processed.csv', index=False)
print('Saved spacex_processed.csv')
```
