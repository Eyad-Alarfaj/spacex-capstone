# IBM Data Science Capstone: SpaceX Falcon 9
## Part 2: Data Wrangling

This notebook performs data cleaning, preprocessing, and feature engineering.

```python3
import pandas as pd
import numpy as np
from datetime import datetime

# Load raw data
df = pd.read_csv('/root/spacex_capstone/data/spacex_launches_raw.csv')
print(f"Loaded {len(df)} records")
print(df.head())
```

## Step 1: Handle Missing Values

```python3
print("Missing Values Before Cleaning:")
print(df.isnull().sum())

# Handle missing Payload_Mass_kg
# Fill with median for missing values
df['Payload_Mass_kg'].fillna(df['Payload_Mass_kg'].median(), inplace=True)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

## Step 2: Feature Engineering - Create Binary Target Variable

```python3
# Create binary target variable 'Class'
# 1 = Successful landing, 0 = Failed landing
df['Class'] = (df['Landing_Outcome'] == 'Success').astype(int)

print("Binary Classification Variable Created:")
print(df[['Landing_Outcome', 'Class']].head(10))

print("\nClass Distribution:")
print(df['Class'].value_counts())
print(f"\nSuccess Rate: {df['Class'].mean():.2%}")
```

## Step 3: Data Type Conversion

```python3
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract additional time features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day_of_Week'] = df['Date'].dt.dayofweek

print("Data types after conversion:")
print(df.dtypes)

print("\nSample with new features:")
print(df[['Date', 'Year', 'Month', 'Day_of_Week', 'Class']].head(10))
```

## Step 4: Feature Encoding - Categorical Variables

```python3
# Create one-hot encoding for categorical features
print("Unique values in categorical columns:")
print(f"Launch Sites: {df['Launch_Site'].nunique()} - {df['Launch_Site'].unique()}")
print(f"Orbits: {df['Orbit'].nunique()} - {df['Orbit'].unique()}")
print(f"Boosters: {df['Booster'].nunique()} - {df['Booster'].unique()}")

# Create dummy variables
df_encoded = pd.get_dummies(df, columns=['Launch_Site', 'Orbit', 'Booster'], drop_first=False)

print(f"\nShape after encoding: {df_encoded.shape}")
print(f"Columns: {list(df_encoded.columns)}")
```

## Step 5: Save Processed Dataset

```python3
# Save both raw features (for visualization) and encoded features (for modeling)

# Simple processed version with key features
df_simple = df[['Flight_Number', 'Date', 'Booster', 'Launch_Site', 'Orbit', 
                  'Payload_Mass_kg', 'Customer', 'Landing_Outcome', 'Year', 
                  'Month', 'Day_of_Week', 'Class']].copy()

df_simple.to_csv('/root/spacex_capstone/data/spacex_launches_processed.csv', index=False)
print("Processed dataset (simple) saved to spacex_launches_processed.csv")

# Encoded version for modeling
df_encoded.to_csv('/root/spacex_capstone/data/spacex_launches_encoded.csv', index=False)
print("Processed dataset (encoded) saved to spacex_launches_encoded.csv")

print(f"\nFinal dataset shape: {df_simple.shape}")
print("\nFirst 5 rows:")
print(df_simple.head())

print("\nData Summary:")
print(df_simple.describe())
```

## Summary

Data wrangling completed successfully:
- Missing values handled (Payload_Mass_kg filled with median)
- Binary target variable created (Class: 1=Success, 0=Failure)
- Temporal features extracted (Year, Month, Day_of_Week)
- Categorical features encoded for modeling
- Two datasets exported: simple (for EDA/visualization) and encoded (for modeling)
