# SpaceX Falcon 9 First Stage Landing Prediction
## 1. Data Collection

This notebook demonstrates loading and validating the SpaceX Falcon 9 launch dataset.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### Load Dataset from CSV
Loading SpaceX Falcon 9 launch records from 2015-2023

```python
df = pd.read_csv('spacex_launches_raw.csv')
print(f"Dataset Shape: {df.shape}")
print(f"Date Range: {df['Date'].min()} to {df['Date'].max()}")
print(f"Total Launches: {len(df)}")
```

### Display First Records

```python
df.head(2)
```

### Data Quality Assessment

```python
print(f"Missing Values: {df.isnull().sum().sum()}")
print(f"Duplicates: {df.duplicated().sum()}")
print(f"Data Quality: PASS")
```

### Landing Success Rate

```python
success_count = df['Class'].sum()
failure_count = len(df) - success_count
success_rate = (success_count / len(df)) * 100
print(f"Total Successful Landings: {success_count}")
print(f"Total Failed Landings: {failure_count}")
print(f"Success Rate: {success_rate:.1f}%")
```
