# SpaceX Falcon 9 First Stage Landing Prediction
## Notebook 1 — Data Collection

Collect and validate Falcon 9 launch records (compiled from the SpaceX REST API `/v4/launches` endpoint and public launch archives) for the landing-prediction capstone.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### Load the launch records

```python
df = pd.read_csv('spacex_launches_raw.csv')
print(f'Dataset shape: {df.shape}')
print(f'Date range: {df.Date.min()} to {df.Date.max()}')
df.head()
```

### Data quality checks

```python
print('Missing values per column:')
print(df.isnull().sum())
print(f'\nDuplicate flights: {df.duplicated().sum()}')
```

### Landing outcome distribution

```python
succ = int(df['Class'].sum()); fail = len(df) - succ
print(f'Successful landings: {succ}')
print(f'Failed landings:     {fail}')
print(f'Success rate:        {succ/len(df)*100:.1f}%')
```

### Summary statistics

```python
print('Payload mass (kg):')
print(df['Payload_Mass_kg'].describe().round(0))
print('\nPrior booster landings:')
print(df['Booster_Landings'].describe().round(1))
```
